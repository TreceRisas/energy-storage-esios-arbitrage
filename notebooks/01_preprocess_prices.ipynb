Depuramos el csv

```python
import os
import pandas as pd

BASE_PATH = "/content/drive/MyDrive/energy_storage_esios"

# Cargar datos raw
raw_path = os.path.join(BASE_PATH, "data", "raw", "esios_price_spot_2024.csv")
df = pd.read_csv(raw_path, parse_dates=["datetime"])

print("Shape original:", df.shape)
print("Geos disponibles:", df["geo_name"].unique())
```

filtramos España

```python
df_es = df[df["geo_name"] == "España"].copy()

print("Shape tras filtrar España:", df_es.shape)
print("Rango:", df_es["datetime"].min(), "->", df_es["datetime"].max())
```

Seleccionamos lo necesario

```python
df_es = df_es[["datetime", "price_eur_mwh"]].sort_values("datetime").reset_index(drop=True)
```

Guardar dataset limpio

```python
processed_dir = os.path.join(BASE_PATH, "data", "processed")
os.makedirs(processed_dir, exist_ok=True)

out_path = os.path.join(processed_dir, "esios_price_spot_es_2024.csv")
df_es.to_csv(out_path, index=False)

print("Guardado:", out_path)
print("Filas finales:", len(df_es))
```
